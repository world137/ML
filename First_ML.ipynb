## Repeat K-Fold validation

```python
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import RepeatedStratifiedKFold
rkf = RepeatedStratifiedKFold(n_splits=3, n_repeats=10, random_state=0)
weight = {0:1.0,1:3.1689}
```

# Modeling

## define evaluation function

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_recall_fscore_support as score
from sklearn.metrics import accuracy_score
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_validate
def getScore(model):
    train_acc = []
    scoring = {'accuracy' : make_scorer(accuracy_score), 
               'precision' : make_scorer(precision_score),
               'recall' : make_scorer(recall_score), 
               'f1_score' : make_scorer(f1_score)}
    
    result = cross_validate(estimator=model,
                                      X=X_scaled,
                                      y=np.ravel(y),
                                      cv=rkf,
                                      scoring=scoring)
    for train_index, test_index in rkf.split(X_scaled,y):
        X_train = X_scaled.iloc[train_index]
        X_test = X_scaled.iloc[test_index]
        y_train = y.iloc[train_index]
        y_test = y.iloc[test_index]
        
        model.fit(X_train,np.ravel(y_train))
        
        y_pred = model.predict(X_test)
        y_pred_train = model.predict(X_train)
        
        train_acc.append(accuracy_score(y_train,y_pred_train))
        
    print("Accuracy from training set =",np.mean(train_acc))
    print("Accuracy from test set =",np.mean(result['test_accuracy']))
    print("Precision =",np.mean(result['test_precision']))
    print("Recall =",np.mean(result['test_recall']))
    print("F1 scores =",np.mean(result['test_f1_score']))
```

## Support Vector Classification

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
params_SVC = {
              'C':[0.1,1,2],
              'gamma':[0.03,0.04,0.05,'auto'], 
              'kernel':['rbf','poly','sigmoid'],
              'class_weight' : [weight],
             }
SVC_grid = GridSearchCV(SVC(),params_SVC,refit=True, verbose=2,n_jobs=-1,cv=rkf,scoring="f1_macro")
SVC_grid.fit(X_scaled,np.ravel(y))
```

```python
SVC_grid.best_params_
```

```python
svc = SVC(C=1,gamma='auto',kernel='rbf',class_weight=weight)
```

```python
getScore(svc)
```

## RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
params_rfm = {
    'n_estimators'      : [50,51,52],
    'max_depth'         : [1,2,3],
    'random_state'      : [0],
    'max_features': ['auto', 'sqrt'],
    'min_samples_split':[2],
    'min_samples_leaf':[4],
    'class_weight' : [weight],
    'criterion' :['gini','entropy']
}
rfm_grid = GridSearchCV(RandomForestClassifier(),params_rfm,refit = True, verbose=2,n_jobs=-1,cv=rkf,scoring='f1')
rfm_grid.fit(X,np.ravel(y))
```

```python
rfm_grid.best_params_
```

```python
rfm = RandomForestClassifier(max_depth= 2, n_estimators=51, random_state= 0,
                              class_weight=weight,min_samples_leaf=4,min_samples_split=2)
getScore(rfm)
```
